```r
y <- cbind(df4$success, df4$failure)

m1 <- glm(y ~ Male + EthnicityWhite + 
            CommitteeId + SubmitterTypeId + IDACIRank,
          family = "binomial", data = df4)

summary(m1)
```

```r
print(plot(m1))
```

```r
m2 <- glm(y ~ Male + EthnicityWhite + 
            CommitteeId + SubmitterTypeId + IDACIRank,
          family = "quasibinomial", data = df4)

summary(m2)
```

```r
m3 <- glm(y ~ Male + EthnicityWhite +
            SubmitterTypeId,
          family = "quasibinomial", data = df4)

# test if simpler model is any different
anova(m2, m3, test = "F")
```

```r
summary(m3)
```

```r
install.packages("FFTrees")
install.packages("caret")
```

```r
library("FFTrees")
library("caret")
```

```r
# train and test split of data randomly from original data
df_parted <- caret::createDataPartition(df_min$Accepted, p = 0.8, list=FALSE)

# Train and test
data_train <- df_min[df_parted, ]
data_test <- df_min[-df_parted, ]
 
#Build the classifier
perf_fft <- FFTrees(formula = Accepted ~.,
                         data = data_train, data.test = data_test)
```

```r
perf_fft
```

```r
plot(perf_fft, 
     main = "Submission acceptance", 
     decision.names = c("Rejected", "Accepted"))
```
