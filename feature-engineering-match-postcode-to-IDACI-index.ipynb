```python
class IdAci(object):
    
    baseUrl = 'https://tools.npeu.ox.ac.uk/imd/'
    
    def __init__(self, postcode):
        self.postcode = postcode
        self.idaciDict = {}

    def connect(self):
        payload['postcode'] = postcode
        try:
            r = requests.post(self.baseUrl, data=payload)
        except ValueError as err:
            print('Get request broken: {}'.format(err))

        output = r.text
        return output  

    def get_idaci(self):
        output = self.connect()
        soup = BeautifulSoup(output, 'lxml')
        spans = [strong_tag.text for strong_tag in soup.find_all('strong')]
        self.idaciDict['idaci_index'] = spans[1]
        self.idaciDict['postcode'] = spans[0]
       
        return self.idaciDict
    
postcode = 0
f'the current postcode is {postcode}'

IdAci(postcode).get_idaci()
    
    
```
